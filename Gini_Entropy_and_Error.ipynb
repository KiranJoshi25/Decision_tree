```
import math
class Decisiontree():
  
  def gini(self,c,d):
    gini_value = 1 - (c/(c+d))**2 - (d/(c+d))**2
    return gini_value
  
  def entropy(self,c,d):
    entropy_value = -(c/(c+d))* math.log2(c/(c+d)) - (d/(c+d))* math.log2(d/(c+d))
    return entropy_value
  
  def classification_error(self,c,d):
    error_value = 1 - max((c/(c+d)),(d/(c+d)))
    return error_value


    
```

```
D = Decisiontree()
```

```
print(D.gini(8,12))
print(D.entropy(8,12))
print(D.classification_error(8,12))
```
